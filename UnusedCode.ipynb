```python
import turtle
import random
import math

# Define constants
WIDTH = 600
HEIGHT = 600
DOT_SIZE_RANGE = (1, 10)
DOT_SPEED_RANGE = (1, 10)
BG_COLOR = "white"
DOT_COLORS = ["red", "green", "blue", "yellow", "cyan", "magenta", "orange", "pink", "brown", "gray"]

# Define Dot class
class Dot:
    def __init__(self, pos):
        self.size = random.randint(*DOT_SIZE_RANGE)
        self.color = random.choice(DOT_COLORS)
        self.speed = random.randint(*DOT_SPEED_RANGE)
        self.direction = random.uniform(0, 2 * math.pi)
        self.pos = pos
    
    def update(self):
        # Move dot based on speed and direction
        dx = self.speed * math.cos(self.direction)
        dy = self.speed * math.sin(self.direction)
        self.pos = (self.pos[0] + dx, self.pos[1] + dy)
        
        # Change direction at random intervals
        if random.random() < 0.1:
            self.direction = random.uniform(0, 2 * math.pi)
    
    def draw(self, t):
        # Draw dot on turtle
        t.penup()
        t.goto(*self.pos)
        t.pendown()
        t.dot(self.size, self.color)

# Define main function
def main():
    # Initialize turtle
    screen = turtle.Screen()
    screen.setup(WIDTH, HEIGHT)
    screen.bgcolor(BG_COLOR)
    screen.title("Game of Life Simulation")
    
    # Create turtle
    t = turtle.Turtle()
    t.hideturtle()
    t.speed(0)
    
    # Create list of dots
    dots = [Dot((random.randint(-WIDTH//2, WIDTH//2), random.randint(-HEIGHT//2, HEIGHT//2))) for i in range(50)]
    
    # Run simulation loop
    running = True
    while running:
        # Update dots
        for dot in dots:
            dot.update()
        
        # Check for collisions
        i = 0
        while i < len(dots):
            j = i + 1
            while j < len(dots):
                if math.dist(dots[i].pos, dots[j].pos) <= dots[i].size + dots[j].size:
                    if dots[i].size > dots[j].size:
                        dots[i].size += 1
                        dots.pop(j)
                        j -= 1
                    else:
                        dots[j].size += 1
                        dots.pop(i)
                        i -= 1
                        break
                j += 1
            i += 1
        
        # Clear turtle
        t.clear()
        
        # Draw dots
        for dot in dots:
            dot.draw(t)
        
        # Update screen
        screen.update()
    
    # Quit turtle
    screen.bye()

# Run main function
if __name__ == "__main__":
    main()
```

```python
import turtle

# Set up the screen
screen = turtle.Screen()
screen.title("Control the Ball")
screen.bgcolor("white")
screen.setup(width=600, height=600)

# Create the ball
ball = turtle.Turtle()
ball.shape("circle")
ball.color("red")
ball.penup()

# Set the ball's movement speed
ball_speed = 10

# Define functions to move the ball using arrow keys
def move_up():
    y = ball.ycor()
    ball.sety(y + ball_speed)

def move_down():
    y = ball.ycor()
    ball.sety(y - ball_speed)

def move_left():
    x = ball.xcor()
    ball.setx(x - ball_speed)

def move_right():
    x = ball.xcor()
    ball.setx(x + ball_speed)

# Bind arrow keys to ball movement functions
screen.onkeypress(move_up, "Up")
screen.onkeypress(move_down, "Down")
screen.onkeypress(move_left, "Left")
screen.onkeypress(move_right, "Right")
screen.listen()

# Keep the screen open until user closes it
turtle.done()
```

```python
import turtle

# Set up the screen
screen = turtle.Screen()
screen.title("Jumping Ball")
screen.bgcolor("white")
screen.setup(width=600, height=600)

# Create the ball
ball = turtle.Turtle()
ball.shape("circle")
ball.color("red")
ball.penup()

# Set the ball's initial position and velocity
ball.goto(0, 250)
ball.dy = 0

# Set the gravity force and jump velocity
gravity = 5
jump_velocity = 5

# Set the ground position and height
ground_pos = -250
ground_height = 20

# Create the ground collider
ground = turtle.Turtle()
ground.penup()
ground.goto(-300, ground_pos)
ground.pendown()
ground.goto(300, ground_pos)
ground.color("green")

# Define the jump function
def jump():
    ball.dy = jump_velocity

# Bind the space bar to the jump function
screen.onkeypress(jump, "space")
screen.listen()

# Define the game loop
def game_loop():
    # Update the ball's position based on its velocity and gravity
    ball.dy -= gravity
    ball.sety(ball.ycor() + ball.dy)
    
    # Check if the ball collides with the ground
    if ball.ycor() <= ground_pos + ground_height:
        ball.sety(ground_pos + ground_height)
        ball.dy = 0
        
    # Update the screen
    screen.update()
    
    # Call the game loop again after a short delay
    screen.ontimer(game_loop, 20)

# Start the game loop
game_loop()

# Keep the screen open until user closes it
turtle.done()
```
